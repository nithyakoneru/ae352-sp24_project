```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
batt_volt = 14.8
batt_cap = 4.2 #Ah
num_motors = 4
amp_draw = 8 # amps drawing from one motor on average
total_runtime = 120 #s
time = np.linspace(0,total_runtime,100)

Energy_batt = batt_volt * batt_cap
power_draw = batt_volt* amp_draw * num_motors

batt_drain = (power_draw/3600) * time
batt_remain = Energy_batt - batt_drain
batt_perc = (batt_remain/Energy_batt)*100

plt.plot(time,batt_perc,'g', linewidth = 3)
plt.axhline(y=20.0, color='r', linewidth=3)
plt.xlabel('Time (s)')
plt.ylabel('Battery Percentage')
plt.xlim([0,total_runtime])
plt.ylim([0,100])
plt.gca().set_yticklabels([f'{i}%' for i in np.arange(0, 101, 20)])
plt.title('Battery Percentage Hover Requirement')
plt.show()
```
